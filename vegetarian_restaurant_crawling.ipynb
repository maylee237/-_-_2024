```python
def execute_select_query(cur, name): #rSTR ID 로 STORE ID 찾는 함수

    # SQL 쿼리 생성
    query = f"""
        SELECT * #,r.ID , r.RSTR_NM,  r.AREA_NM_1 , r.AREA_NM_2,r.AREA_NM_3, r.RSTR_RDNMADR
        FROM RSTR_TRANSLATE r
        WHERE r.RSTR_NM LIKE "{name}%"
        AND r.LANG_NM = 'EN'
    """

    # 쿼리 실행
    cur.execute(query)

    re = cur.fetchall()
    
    if not re:  # 결과값이 없는 경우
        # 널값을 추가하여 데이터프레임 생성
        columns = [desc[0] for desc in cur.description]
        df = pd.DataFrame(columns=columns)
        df.loc[0] = [None] * len(columns)
    else:
        # 결과값을 데이터프레임으로 변환
        columns = [desc[0] for desc in cur.description]
        df = pd.DataFrame(re, columns=columns)

    # 결과 반환
    return df

# 사용 예시
# cur: 데이터베이스 커서 객체
# integrated_ids: 조회할 integrated_id 리스트
# results = execute_select_query(cur, [141438, 392054, 38918, 4601])
```

```python
from fuzzywuzzy import fuzz

def address_match_ratio(address1, address2):
    if pd.notna(address1) and pd.notna(address2):  # NaN이 아닌 경우에만 처리
        return fuzz.partial_ratio(address1, address2)
    else:
        return 0
    
# 두 열 간의 값 비교 및 유사성 계산
tr['similarity_ratio2'] = tr.apply(lambda row: address_match_ratio(row['RDNMADR'], row['Address']), axis=1)
tr
```

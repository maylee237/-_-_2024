### 서울 상점 링크 크롤링

```python

# Selenium 웹드라이버 시작
driver = webdriver.Chrome()

# https://guide.michelin.com/kr/en/seoul-capital-area/kr-seoul/restaurants/page/10?sort=distance 
# 해당 사이트 접속후 전체 페이지수대로 range 진행

store_link = []
for i in range(1):    
    driver.get(f"https://guide.michelin.com/kr/en/seoul-capital-area/kr-seoul/restaurants/page/{i}?sort=distance")
    html_content = driver.page_source
    soup_1 = BeautifulSoup(html_content, 'html.parser')
    # 'div' 태그 중에서 class가 'flex_fill'인 요소를 찾습니다.
    for href in soup_1.find_all('div', class_='col-md-6'):
        print(href.find("a")["href"])
        store_link.append(href.find("a")["href"])
        # 다 찾으면 옆으로 가기 버튼을 누르면서 하나씩 이동


driver.quit()
```

### 부산 상점 링크 크롤링

```python

# Selenium 웹드라이버 시작
driver = webdriver.Chrome()

busan_store_link = []
for i in range(4):     
    driver.get(f"https://guide.michelin.com/kr/en/busan-region/restaurants/page/{i}")
    html_content = driver.page_source
    soup_1 = BeautifulSoup(html_content, 'html.parser')
    # 'div' 태그 중에서 class가 'flex_fill'인 요소를 찾습니다.
    for href in soup_1.find_all('div', class_='col-md-6'):
        print(href.find("a")["href"])
        busan_store_link.append(href.find("a")["href"])
        # 다 찾으면 옆으로 가기 버튼을 누르면서 하나씩 이동


driver.quit()
```

### 2. 추출한 링크 이용하여 세부 내용 추출

```python
link =  store_link + busan_store_link
```

```python
title_list = []
address_list = []
price_list = []
classfication_list = []
description_list = []
phone_list = []
price=[]
label=[]
postcode=[]
urls = []

for i in store_link:
    url = "https://guide.michelin.com/" + i
    urls.append(url)
    # HTML 내용을 가져옴
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, 'html.parser')
    time.sleep(1)

    # 제목 가져오기
    title = soup.find('h1', class_='restaurant-details__heading--title').text.strip()
    title_list.append(title)
    print("Title:", title)

    # 주소 및 가격 정보 가져오기
    address_info = soup.find('li', class_='restaurant-details__heading--address').text.strip()
    address_list.append(address_info)
    print("Address:", address_info)

    postcode.append(address_info.split(',')[-2]) #영어는 , 기준 -2

    price_info = soup.find('li', class_='restaurant-details__heading-price').text.strip()
    price_list.append(price_info.strip().split())
    #금액이랑 라벨 나누기
    print("Price !! :", price_info.strip().split())
    price.append(price_info.strip().split()[0]) #영어는 나눠서 0번째
    label.append(price_info.strip().split()[2])


    # 분류 가져오기
    classfication_element = soup.find('div', class_='restaurant-details__classification')
    if classfication_element:
        classfication = classfication_element.text
        classfication = re.sub(r'\s+', ' ', classfication)
        classfication_list.append(classfication)
        print("분류:", classfication)
    else:
        classfication_list.append(None)
        print("분류: 정보 없음")
      

    # 설명 가져오기
    des = soup.find('div', class_='restaurant-details__description--text').text
    description_list.append(des)
    print("설명:", des)

    # 전화번호 가져오기
    phone = soup.find('span', class_='flex-fill').text
    phone = re.sub(r'\s+', ' ', phone)
    phone_list.append(phone)
    print("전화:", phone)

    print()
```

```python
# 데이터프레임 생성
seoul = {
    'Title': title_list,
    'Address': address_list,
    '우편번호' : postcode,
    '가격' : price,
    'label' : label,
    '분류': classfication_list,
    '설명': description_list,
    '전화': phone_list
}

seoul = pd.DataFrame(seoul)
seoul
```

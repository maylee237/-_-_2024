```python
# 매칭후 정확도를 위해 유사성 계산 진행

from fuzzywuzzy import fuzz

def address_match_ratio(address1, address2):
    if pd.notna(address1) and pd.notna(address2):  # NaN이 아닌 경우에만 처리
        return fuzz.partial_ratio(address1, address2)
    else:
        return 0
    
# 두 열 간의 값 비교 및 유사성 계산
merge3['similarity_ratio'] = merge3.apply(lambda row: address_match_ratio(row['상점명'], row['name']), axis=1)
merge3
```
